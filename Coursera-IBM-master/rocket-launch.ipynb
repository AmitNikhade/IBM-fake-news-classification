Feature Engineering and Selection

```python
df['DateTime'] = pd.to_datetime(df['Datum'])

df['Year'] = df['DateTime'].apply(lambda datetime: datetime.year)

df["Country"] = df["Location"].apply(lambda location: location.split(", ")[-1])
```

```python
enc=LabelEncoder()
df['Company Name'] = enc.fit_transform(df['Company Name'])
df['Status Rocket'] = enc.fit_transform(df['Status Rocket'])
df['Country'] = enc.fit_transform(df['Country'])
```

```python
sns.pairplot(df)
```

```python
df = df.drop(columns = ['Unnamed: 0','Unnamed: 0.1','Datum','Location', 'Detail', 'DateTime',' Rocket'], axis = 1)
```

```python
df
```

```python
X= df[['Company Name', 'Status Rocket', 'Year', 'Country']]
y= df['Status Mission']
```

Splitting data into test and train

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

Classifier model

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=10)
clf1 = AdaBoostClassifier(n_estimators=50, learning_rate=0.001)
clf2 = SGDClassifier(random_state=None)

clf.fit(X_train, y_train)
prediction_rf=clf.predict(X_test)
print(confusion_matrix(y_test,prediction_rf))
print('\n')
print(classification_report(y_test,prediction_rf))


clf1.fit(X_train, y_train)
prediction_rf=clf1.predict(X_test)
print(confusion_matrix(y_test,prediction_rf))
print('\n')
print(classification_report(y_test,prediction_rf))


clf2.fit(X_train, y_train)
prediction_rf=clf2.predict(X_test)
print(confusion_matrix(y_test,prediction_rf))
print('\n')
print(classification_report(y_test,prediction_rf))
```

```python
rf=RandomForestClassifier()
```

```python
parameters = {'n_estimators':[50,100,150,200,250,300,350,400],'criterion':('gini', 'entropy'), 'random_state':[1, 10, 20, 30 ,40, 50, 60], 'max_features':('auto', 'sqrt'), 'min_samples_leaf':[1, 2, 4], 'bootstrap':[True, False]}
clf = GridSearchCV(rf, parameters, cv = 3, verbose=2,  n_jobs = -1)
clf.fit(X_train,y_train)
```

Predictions

```python
clf.best_params_
```

```python
rf_model=RandomForestClassifier(n_estimators=150, criterion='entropy', min_samples_leaf=4, max_features='auto', bootstrap=True, n_jobs=-1, random_state=10, verbose=0)
rf_model.fit(X_train,y_train)
```

```python
prediction_rf=rf_model.predict(X_test)
```

```python
rf_model.score(X_test, y_test)
```

We have got a good score without overfitting
